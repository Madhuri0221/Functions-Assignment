```python
# 1) Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list

def sum_of_even_numbers(l):
  '''
  Args: List of numbers is taken as input
  Returns: Returns sum of all even numbers in the given range

  '''
  s=0
  for i in l:
    if i%2 == 0:
      s+=i
    print(i,end=" ")
  return s
sum_of_even_numbers([i for i in range(20)])
```

```python
# 2) Create a Python function that accepts a string and returns the reverse of that string

def reverse_of_string(s):

  '''
  Args: A string is taken as input"
  Returns: Returns reverse of that string

  '''
  t=""
  for c in s:
    t=c+t
  return t
reverse_of_string("i like japan")
```

```python
# 3) Implement a Python function that takes a list of integers and returns a new list containing the squares of each number

def squares_of_numbers(l):

  '''
  Args: Tkes list as input
  Returns: Return square of each number of the list in the new list

  '''
  li=[i**2 for i in l]
  return li
squares_of_numbers([i for i in range(5)])

```

```python
# 4) Write a Python function that checks if a given number is prime or not from 1 to 200

def prime_number(num):

  '''
  Args: Takes a number as input
  Returns: Returns if the given number is prime or not

  '''
  for i in range(2,int(num**0.5)+1):
    if num%i == 0:
      return False
  else:
      return True
print(prime_number(3))

for n in range(1,11):
  if prime_number(n):
    print(f"{n} is prime number")
  else:
    print(f"{n} is not prime number")
```

```python
# 6) Write a generator function in Python that yields the powers of 2 up to a given exponent

def power_of_number(base,exponent):
  yield [base**i for i in range(exponent+1)]
gen = power_of_number(2,5)
print(next(gen))

```

```python
# 10) Create a Python program that uses `filter()` to remove all the vowels from a given string

def filter_vowels(s):
  v="AEIOUaeiou"
  for c in s:
    if c not in v:
      return c
# filter_vowels("mango")

def string(s):
  return ''.join(filter(filter_vowels,s))
string("earth is a planet")
```

```python
# 7)  Implement a generator function that reads a file line by line and yields each line as a string.

def read_lines(filepath):
  with open(filepath,'r') as file:
    for line in file:
      yield line
filepath = "D:\Madhuri\PW_Skills\Water_bodies.txt"
gen = read_lines(filepath)
print(next(gen),next(gen),next(gen),next(gen))
```

```python
# 6. Explain the concept of generators in Python and how they are defined.

'''
generators yield values one-by-one instead of returning them all at once (lazy evaluation)
these are deined using yield instead of return
yield pauses the execution
return terminate the execution
saves memory
we can create an object for the generator function and by using next() we can get each value


def func(t):
  yield t
gen = func((1,2,3,4,5,6))
print(next(gen),next(gen)) -> gives 1,2 as output

'''
```

```python
# 9. Explain the purpose and usage of the `map()` function in Python.

'''
map() syntax: map(func,iterable)
ie., we are mapping the function to the iterable
the function may be a custom or built function or even we can use lambda function

Applying single function uniformly across items
'''
l=['1','2','3']
int_l=map(int,l)
print(list(int_l))

result = list(map(lambda x: x+2,[i for i in range(5)]))
print(result)

t1=(1,2,3,4,5)
t2=(1,2,3,4,5)

result = list(map(lambda x,y: x*y,t1,t2))
print(result)

```

```python
# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

'''
map(),reduce(),filter() all three take function as first argument and 2nd is the iterable

map(): this applies the function uniformly to the items of the iterable

reduce(): this reduces the list of items into a single value as output

filter(): this filters and gives the desired output based on some conditions

'''

# map():
result = list(map(lambda x: x+2,[i for i in range(5)]))
print(result)

# filter():
l=[1,2,-9,-8]
result = list(filter(lambda x: x if x>0 else 0,l))
print(result)

# reduce()
from functools import reduce
l=[1,2,-1,-2,10]
result = reduce(lambda x,y:x+y,l)
print(result)
```
